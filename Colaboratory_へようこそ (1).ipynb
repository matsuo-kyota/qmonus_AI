## テキスト要約API

```
from transformers import pipeline

summarizer = pipeline("summarization")

def summarize_text(text, max_length=100):
  """
  与えられたテキストを要約します。

  Args:
      text (str): 要約するテキスト。
      max_length (int): 要約の最大文字数。デフォルトは100。

  Returns:
      str: 要約されたテキスト。
  """
  summary = summarizer(text, max_length=max_length, min_length=30)[0]['summary_text']
  return summary

if __name__ == "__main__":
  text = "これは長い文章です。要約して欲しいです。要約は短く、簡潔に。重要な部分を抜き出して記述してください。"
  summary = summarize_text(text)
  print(summary)
```

## 感情分析API

```
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

def analyze_sentiment(text):
  """
  与えられたテキストの感情を分析します。

  Args:
      text (str): 分析するテキスト。

  Returns:
      dict: 感情分析結果 (label: 感情ラベル, score: 感情スコア)。
  """
  result = classifier(text)[0]
  return {"label": result['label'], "score": result['score']}

if __name__ == "__main__":
  text = "今日はとても良い天気です。"
  sentiment = analyze_sentiment(text)
  print(sentiment)
```

## 分類(ゼロショット分類)

```
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-xnli-multilingual-nli-2mil7")
classifier(
    "明日の天気は晴れ。",
    candidate_labels=["気象", "政治", "気分"],
)
```
